```python
# 4. Сколько у каждого из пользователей в среднем покупок в неделю (по месяцам)?
```

```python
# При определении среднего количества покупок у каждого пользователя по месяцам считаю целесообразным учитывать только
# заказы со статусом "доставлен пользователю", т.к. некорректно учитывать иные статусы - заказы, которые еще не завершены и 
# могут отмениться и т.п. Иными словами, пользователь может оформлять большое количество заказов,
# но отменять их или отказываться при получении, и считать такие заказы совершенными покупками неверно - денег они не приносят.
# Таким образом, я буду учитывать только заказы со статусом "доставлен пользователю"
# за исключением 8 заказов с отсутствующими датами доставки, но со статусом "доставлен" (причину указывала в заданиях выше).
```

```python
# Отбираю нужные данные:
avg_orders_by_weeks = customers.merge(orders, how='inner', on='customer_id') \
                               .query('order_status == "delivered" and order_delivered_customer_date != "NaT"') 
avg_orders_by_weeks
```

```python
# Определяю количество недель в каждом месяце:
avg_orders_by_weeks['amount_of_weeks'] = (avg_orders_by_weeks.order_purchase_timestamp.dt.daysinmonth / 7).round(2)
avg_orders_by_weeks
```

```python
# Перевожу дату заказа в нужный формат и определяю общее количество заказов в разрезе пользователей и месяцев:
avg_orders_by_weeks['order_purchase_timestamp'] = avg_orders_by_weeks['order_purchase_timestamp'].dt.strftime("%Y-%m")
avg_orders_by_weeks = avg_orders_by_weeks.groupby(['customer_unique_id', 'order_purchase_timestamp', 'amount_of_weeks'], as_index=False) \
                                         .agg({'order_id' : 'count'}) \
                                         .rename(columns={'order_id':'orders', 'order_purchase_timestamp':'month'}) 
avg_orders_by_weeks
```

```python
# Определяю среднее количество покупок по месяцам в разрезе пользователей:
avg_orders_by_weeks['avg_orders_by_weeks'] = (avg_orders_by_weeks.orders / avg_orders_by_weeks.amount_of_weeks)

avg_orders_by_weeks[['customer_unique_id', 'month', 'amount_of_weeks', 'avg_orders_by_weeks']]
```

```python
# 5. Используя pandas, проведи когортный анализ пользователей.
# В период с января по декабрь выяви когорту с самым высоким retention на 3й месяц.
```

```python
# Для решения задачи я буду учитывать только заказы со статусом "доставлен пользователю",
# за исключением 8 заказов с отсутствующими датами доставки, но со статусом "доставлен" (причину указывала в задачах выше).
# Для составления когорт я буду брать данные только за 2017 год, т.к. данные за 2016 и 2018 годы неполные (не за все месяцы).
```

```python
# Отбираю нужные данные (доставленные заказы) и перевожу дату заказа в нужный формат:
delivered_orders = customers.merge(orders, how='inner', on='customer_id') \
                               .query('order_status == "delivered" and order_delivered_customer_date != "NaT"') 
delivered_orders['order_purchase_timestamp'] = delivered_orders['order_purchase_timestamp'].dt.strftime("%Y-%m") 
delivered_orders
```

```python
# Определяю дату первого заказа пользователя (т.е. когорту, к которой он относится):
first_orders = delivered_orders.groupby('customer_unique_id', as_index=False) \
                               .agg({'order_purchase_timestamp': 'min'}) \
                               .rename(columns = {'order_purchase_timestamp': 'cohort'})
delivered_orders_cohorts = delivered_orders.merge(first_orders, how='left',on='customer_unique_id')
delivered_orders_cohorts
```

```python
# Отбираю необходимые данные за 2017 год:
delivered_orders_cohorts = delivered_orders_cohorts.query("'2017-01' <= cohort <= '2017-12'") \
                                                   .groupby(['cohort', 'order_purchase_timestamp'], as_index=False) \
                                                   .agg({'customer_unique_id': 'count'}) \
                                                   .rename(columns={'order_id':'orders', 'customer_unique_id':'clients_amount'})
delivered_orders_cohorts
```

```python
# Считаю количество пользователей в каждой когорте:
amount_in_cohorts = delivered_orders_cohorts.groupby('cohort', as_index=False) \
                                            .agg({'clients_amount': 'max'}) \
                                            .rename(columns = {'clients_amount': 'amount_in_cohort'})
delivered_orders_cohorts = delivered_orders_cohorts.merge(amount_in_cohorts, how = 'left', on = 'cohort')  
delivered_orders_cohorts
```

```python
# Считаю retention:
delivered_orders_cohorts['retention'] = delivered_orders_cohorts['clients_amount'] / delivered_orders_cohorts['amount_in_cohort']
delivered_orders_cohorts
```

```python
# Формирую сводную таблицу:
retention = delivered_orders_cohorts.pivot_table(index = 'order_purchase_timestamp', columns = 'cohort', values = 'retention')
retention
```

```python
sns.set (rc={"figure.figsize": (25, 8)})
sns.heatmap(retention, vmin=0.0000000, vmax = 0.005, annot=True)
```

```python
# 6. Часто для качественного анализа аудитории использую подходы, основанные на сегментации.
# Используя python, построй RFM-сегментацию пользователей, чтобы качественно оценить свою аудиторию.
# В кластеризации можешь выбрать следующие метрики: R - время от последней покупки пользователя до текущей даты,
# F - суммарное количество покупок у пользователя за всё время, M - сумма покупок за всё время.
# Подробно опиши, как ты создавал кластеры. Для каждого RFM-сегмента построй границы метрик recency, frequency и monetary
# для интерпретации этих кластеров. Пример такого описания: RFM-сегмент 132 (recency=1, frequency=3, monetary=2)
# имеет границы метрик recency от 130 до 500 дней, frequency от 2 до 5 заказов в неделю,
# monetary от 1780 до 3560 рублей в неделю. 
```

```python
# Для решения задачи я буду учитывать только заказы со статусом "доставлен покупателю",
# за исключением 8 заказов с отсутствующими датами доставки, но со статусом "доставлен" (причину указывала в задачах выше).
```

```python
# Отбираю нужные данные путем объединения исходных датафреймов:
df_RFM = customers.merge(orders, how='inner', on='customer_id') \
                  .query('order_status == "delivered" and order_delivered_customer_date != "NaT"') 
df_RFM = df_RFM.merge(items, how='inner', on='order_id')
df_RFM
```

```python
# Оставляю только необходимые столбцы с данными:
df_RFM = df_RFM .groupby(['order_id', 'customer_unique_id', 'order_purchase_timestamp'], as_index=False) \
                .agg({'price': 'sum'})
df_RFM
```

```python
# Для определения времени, прошедшего с момента последней покупки покупателя до текущей даты, в качестве текущей даты
# я считаю правильным принять дату следующего дня с момента последней покупки среди всех покупателей по имеющимся данным:
current_date = (df_RFM['order_purchase_timestamp'].max() + datetime.timedelta(days=1)).date()
current_date
```

```python
# Привожу столбец с датами заказов к виду "год-день-месяц":
df_RFM.order_purchase_timestamp = pd.to_datetime(df_RFM.order_purchase_timestamp).dt.date
df_RFM
```

```python
# Считаю время в днях от последней покупки покупателя до текущей даты:
df_RFM['delta_time'] = (current_date - df_RFM.order_purchase_timestamp).dt.days
df_RFM
```

```python
# Создаю RFM-таблицу:
RFM = df_RFM.groupby('customer_unique_id', as_index=False) \
            .agg({'delta_time': 'min', 
                  'order_id':'nunique', 
                  'price': 'sum'}) \
            .rename(columns={'delta_time': 'Recency', 'order_id': 'Frequency', 'price': 'Monetary'})
RFM
```

```python
# Далее я буду использовать перцентили и разделю при помощи них покупателейй на 3 группы в разрезе каждого параметра,
# после чего присвою покупателям оценку по каждому параметру, где: 1 - наихудшее значение, 3 - наилучшее значение.
```

```python
# Определяю перцентили по каждому из параметров:
percentiles = RFM.quantile([0.25, 0.75])
percentiles
```

```python
# Очевидно, что для параметра Frequency данная разбивка не подходит, поэтому сгруппирую покупателей по количеству заказов:
RFM.groupby('Frequency').agg({'customer_unique_id' : 'count'})
```

```python
# На основании полученных данных приму следующее деление покупателей на группы по параметру Frequency:
# 1 - 1 покупка, 2 - 2 покупки, 3 - 3 и более покупок.
# Деление покупателей на группы по параметру Recency (исходя из значений перцентилей):
# 1 - более 347 дней, 2 - от 116 до 347 дней, 3 - 115 и менее дней.
# Деление покупателей на группы по параметру Monetary (исходя из значений перцентилей):
# 1 - 47,65 и менее, 2 - от 47,66 до 154,7, 3 - более 154,7.
```

```python
# Составляю функции для разбивки значений параметров по 3 группам (1, 2, 3):
def r_score(x):
    if x > 347:
        return 1
    elif 116 <= x <= 347:
        return 2
    elif x <= 115:
        return 3
    
def f_score(x):
    if x == 1:
        return 1
    elif x == 2:
        return 2
    elif x > 2:
        return 3

def m_score(x):
    if x <= 47.65:
        return 1
    elif 47.66 <= x <= 154.7:
        return 2
    elif x > 154.7:
        return 3
```

```python
# Оцениваю каждого покупателя по 3 параметрам при помощи созданных выше функций:
RFM['R'] = RFM.Recency.apply(lambda x: r_score(x))
RFM['F'] = RFM.	Frequency.apply(lambda x: f_score(x))
RFM['M'] = RFM.Monetary.apply(lambda x: m_score(x))
RFM
```

```python
# Объединяю оценки по каждому параметру покупателя в общий сегмент RFM: 
RFM['RFM'] = RFM['R'].map(str) + RFM['F'].map(str) + RFM['M'].map(str)
RFM
```

```python
# Н., клиенты с сегментом 333 - это клиенты, которые делали заказы недавно (последний заказ был не более 4 месяцев назад),
# за все время совершили не менее 3 покупок и сумма заказов каждого из них превышает сумму заказов 75% покупателей.
# Определю количество пользователей с таким сегментом:
RFM.query('RFM == "333"').customer_unique_id.count()
```

```python
# Клиенты с сегментом 111 - это клиенты, которые совершили заказ давно(последний и единственный их заказ был около года
# назад или более), они сделали всего одну покупку и сумма заказов каждого из них меньше суммы заказов 75% покупателей. 
# Определю количество пользователей с таким сегментом:
RFM.query('RFM == "111"').customer_unique_id.count()
```

```python
# Определю количество покупателей посегментно и процент каждого сегмента относительно общего числа покупателей:
RFM = RFM.groupby('RFM', as_index=False).agg({'customer_unique_id' : 'count'}) \
                                  .rename(columns={'customer_unique_id':'amount_of_klients'})
RFM['percent'] = (RFM.amount_of_klients * 100 / RFM.amount_of_klients.sum()).round(3)
RFM.sort_values('percent', ascending=False)
```

```python
sns.set(
    font_scale=1,
    style="whitegrid"
        )
plt.figure(figsize=(16,13))
plt.title('RFM')
sns.barplot(x='RFM', y='percent', data=RFM)
```
